Perhatikan kolom `Date Received, Incident Date, Claim Amount` dan `Close Amount`. Ubah data tipe di kolom-kolom tersebut sesuai dengan isinya. (cek to_datetime)

```python
df['Date Received'] = pd.to_datetime(df['Date Received'])
df['Incident Date'] = pd.to_datetime(df['Incident Date'])
```

```python
df[['Claim Amount','Close Amount']] = df[['Claim Amount','Close Amount']].replace('[\$,]', '', regex=True).replace('[;,]','', regex=True).replace('-','NaN', regex=True).astype(float)
# df['Claim Amount'] = df['Claim Amount'].str.replace('$','').str.replace(';','').apply(lambda x: np.nan if x =='-' else x).astype(float)
```

Kita ingin menganalisa berapa lama pelanggan melaporkan kehilangan barang. Buatlah kolom baru bernama `Day Differences`, yang merupakan selisih dari  `Date Received` dan `Incident Date`. (perhatikan data tipenya)

```python
df['Day Differences'] = (df['Date Received'] - df['Incident Date']).dt.days
df['Day Differences']
```
