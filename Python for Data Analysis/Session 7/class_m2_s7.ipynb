## Multi Index

```python
# create multi index
# index level
outside='Jakarta Jakarta Jakarta Surabaya Surabaya Surabaya'.split()
inside=[1,2,3,1,2,3]
hier_index=list(zip(outside,inside))
hier_index
```

```python
hier_index=pd.MultiIndex.from_tuples(hier_index) #data tipe menjadi multi index untuk pandas dataframe
hier_index 
```

```python
df_multi=pd.DataFrame(np.random.randint(1,100,12).reshape(6,2), columns='Smartphone PC'.split(), index=hier_index)
df_multi
```

```python
df_multi.loc['Jakarta'] #menampilkan data semua toko di kota Jakarta
# multi index memudahkan kita dalam pencarian data kelompok tertentu
```

```python
df_multi.loc['Jakarta'].loc[2] #menampilkan data penjualan toko 2 di kota jakarta
```

```python
df_multi.xs(['Jakarta',2]) #pada multiindex, kita bisa melakukan selection dengan fungsi xs
```

```python
df_multi.loc[('Jakarta',2)] # dengan loc, pakai tuple untuk memanggil 2 index dengan hiararki berbeda
```

```python
df_multi
```

```python
df_multi.index.names #index df_multi belum memiliki nama
```

```python
df_multi.index.names=['Kota','Lokasi']
df_multi
```

```python
df_multi.xs(1,level='Lokasi') #memanggil toko-toko di lokasi 1 dari tiap-tiap kota
```

## Missing Value
data kosong, yang biasanya ditandai dengan NaN (nont a number)

```python
df.isna() #cek apakah suatu data adalah NaN atau bukan
```

```python
df.isna().sum() #menampilkan ada berapa data NaN pada suatu kolom
```

```python
df.isnull().sum()
```

```python
df_contoh = pd.DataFrame({
    'Jakarta' : [100, 200, np.nan],
    'Bandung' : [500, np.nan, np.nan],
    'Bekasi' : [100, 200, 300]
})
df_contoh
```

```python
df_contoh.dropna() #secara default, akan menghapus SEMUA BARIS yang memiliki data NaN
```

```python
df_contoh.dropna(axis=1) # menghapus SEMUA KOLOM yang memiliki data NaN
```

```python
df_contoh.dropna(subset=['Jakarta',]) #menghapus baris dengan data NaN berdasarkan kolom Jakarta 
```

```python
df_contoh.dropna(thresh=2) #menghapus baris yang memiliki data NaN lebih dari 2
```

```python
df_contoh.fillna(value='Missing Data') #untuk mengisi NaN dengan value tertentu
```

```python
df_contoh['Jakarta'].fillna(df_contoh['Jakarta'].mean()) 
#mengisi data NaN pada kolom Jakarta dengan rata-rata nilai kolom Jakarta
# jangan lupa disimpan/inplace
```

## Agregasi

```python
df.groupby('Year').std()
```

```python
df.groupby('Year').max()['TotalPayBenefits'] #nilai total pay benefit paling tinggi di tiap tahun
```

```python
df.groupby('Year',as_index=False).mean() #as index=false agar kolom yang dijadikan group tidak menjadi index
```
