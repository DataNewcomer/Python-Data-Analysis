# **Import Library**

```python
import pandas as pd
import numpy as np
```

# **Baca file csv loan_dem.csv dan loan_fina.csv sebagai df1 dan df2**

```python
df1=pd.read_csv('loan_dem.csv')
df2=pd.read_csv('loan_fina.csv')
```

**Cek 5 data teratas df1 dan df2**

```python
df1.head(5)
```

```python
df2.head(5)
```

**Perhatikan bahwa kedua dataframe memiliki kolom `unnamed: 0`. Drop kolom tersebut**

```python
df1.drop('Unnamed: 0', axis=1, inplace=True)
df1
```

```python
df2.drop('Unnamed: 0', axis=1, inplace=True)
df2
```

**Kedua dataframe, sebenarnya merupakan data demografi dan finansial dari aplikan yang sama (perhatikan Loan_ID). Karenanya, gabungkan kedua dataframe tersebut menjadi 1 dataframe di variabel df, lalu tampilkan 5 data paling atas.**

```python
df_merge = pd.merge(df1, df2, on='Loan_ID')
df_merge.head()
```

**Sekarang, perhatikan kolom `Gender` dan kolom  `Property_Area`, gantilah isi dari kedua kolom tersebut dengan ketentuan berikut:**
* **Pada kolom `Gender`, gantilah angka 1 menjadi `Male` dan angka 0 menjadi `Female`**
* **Pada kolom `Property_Area`, gantilah angka 1 menjadi `Semiurban`, 2 menjadi `Urban`, dan 3 menjadi `Rural`**


```python
df_merge['Gender']=np.where(df_merge['Gender']==1, 'Male', 'Female')
df_merge['Property_Area']=df_merge['Property_Area'].apply(lambda x: 'Semiurban' if x==1 else 'Urban' if x==2 else 'Rural')
```

```python
df_merge[['Gender', 'Property_Area']]
```

**Tampilkan juga jumlah missing value dari masing-masing kolom**

```python
df_merge.isna().sum()
```

**Tampilkan data yang berisikan Nan pada kolom `LoanAmount`. Simpan index subset ini ke dalam variabel index_nan**

```python
df_merge.loc[df_merge['LoanAmount'].isna()]
```

```python
index_nan = df_merge.loc[df_merge['LoanAmount'].isna()].index
index_nan     #index dari baris-baris yang loan amountnya kosong
```

**Sekarang, coba cek rata-rata `LoanAmount` berdasarkan kolom `Loan_Status`**

```python
df_merge.groupby('Loan_Status').mean()['LoanAmount']
```

Sekarang, isi data kosong di kolom `LoanAmount` dengan **rata-rata `LoanAmount` berdasarkan kolom `Loan_Status`**.  

Jadi, jika kolom `Loan_Status` berisi `Y`, maka pengganti Nan di kolom `LoanAmount` adalah `144.294404`, Jika `N`, maka `151.220994`.  Cek, apakah data sudah terisi dengan benar. 

```python
df_merge.groupby('Loan_Status')['LoanAmount'].apply(lambda x: x.fillna(x.mean()))
```

```python
df_merge['LoanAmount']= df_merge.groupby('Loan_Status')['LoanAmount'].apply(lambda x: x.fillna(x.mean()))
```

```python
df_merge.loc[index_nan,['LoanAmount', 'Loan_Status']]
```

**Dari informasi di atas, kita bisa melihat bahwa kolom `Dependents` berisikan data tipe object. Coba periksa isi dari kolom `Dependents`**

```python
df_merge['Dependents'].value_counts()
```

**Kita ingin agar kolom `Dependent` memiliki data tipe numeric. Ganti value 3+ menjadi 3, lalu ganti tipe data kolom `Dependent` menjadi numeric** (Belum Selesai)

```python
df_merge['Dependents']= df_merge['Dependents'].replace('3+',3)
df_merge['Dependents']=pd.to_numeric(df_merge['Dependents'])
df_merge['Dependents']
```

```python
df_merge['Dependents'].isna().sum()
```

**Sekarang, buatlah kolom baru `TotalIncome` yang berisi jumlah dari `ApplicantIncome` dan `CoapplicantIncome`. Lalu buat kolom lain bernama `IncomeLoanRatio` yang berisikan ratio antara `TotalIncome` dan `LoanAmount`**

```python
df_merge['TotalIncome'] = df_merge['ApplicantIncome']+df_merge['CoapplicantIncome']
df_merge['IncomeLoanRatio']=(df_merge['TotalIncome']/df_merge['LoanAmount']).round(2)
df_merge.head()
```
