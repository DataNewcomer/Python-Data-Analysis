# **Exercise: Pandas Merging**

Dataset: audi.csv, hyundai.csv, toyota.csv

```python
import numpy as np
import pandas as pd
```

Dataset: https://www.kaggle.com/adityadesai13/used-car-dataset-ford-and-mercedes?select=toyota.csv

```python
df_audi = pd.read_csv('audi.csv')
df_audi.head()
```

```python
df_hyundai = pd.read_csv('hyundai.csv')
df_hyundai.head()
```

```python
df_toyota = pd.read_csv('toyota.csv')
df_toyota.head()
```

## **No.1**

```python
# Jika Anda perhatikan, nama kolom 'tax(£)' di dataset hyundai berbeda dengan di dataset lainnya.
# Ubah nama kolom tersebut menjadi 'tax' agar sama dengan di dataset yang lain.

df_hyundai.rename(columns={'tax(£)':'tax'}, inplace=True)
df_hyundai.head(3)
```

## **No.2**

Buat kolom baru bernama ‘brand’ dengan isi nama brand mobilnya, letakkan pada kolom paling kiri. Misalnya untuk data frame audi, isi kolom ‘brand’ dengan ‘Audi’!

```python
# insert(ColumnIndex, ColumnName, Value)

df_audi.insert(0, 'brand', value='Audi')
df_hyundai.insert(0, 'brand', value='Hyundai')
df_toyota.insert(0, 'brand', value='Toyota')
```

```python
df_toyota.head(3)
```

## **No.3**

Gabungkan ketiga data frame tersebut menjadi sebuah data frame bernama df_car. Atur agar indexnya tidak ada yang berulang.

```python
df_car = pd.concat([df_audi,df_hyundai,df_toyota], axis=0, ignore_index=True)
df_car
```

## **No.4**

Pada kolom ‘model’ di sebagian data, ada whitespace tambahan di awal atau di akhir yang tidak berguna. Hapus whitespace tersebut dari semua data pada kolom ‘model’!

```python
# padahal pada index 0, tertulis 'A1' tetapi len-nya 3. Berarti ada whitespace
df_car['model'].apply(lambda x: len(x))[:5]
```

```python
# menghilangkan whitespace
df_car['model'] = df_car['model'].str.strip()
```

```python
# sekarang udah benar, 'A1' len-nya adalah 2
df_car['model'].apply(lambda x: len(x))[:5]
```

## **No.5**

Dari df_car, tampilkan rata-rata pajak/tax berdasarkan brand dan transmission. Lalu urutkan dari tax yang paling tinggi


```python
df_tax = df_car.groupby(by=['brand','transmission'], as_index=False).mean()[['brand','transmission','tax']].sort_values(by='tax', ascending=False)
df_tax
```

```python
df_tax.pivot_table(index='brand', columns='transmission', values='tax')
```

## **No.6**

Cari mobil Toyota Camry yang harganya (price) di bawah rata-rata harga Camry dan jarak tempuhnya (mileage) di bawah rata-rata mileage Camry juga. Tampilkan mobil yang sesuai dengan kriteria tersebut!


# Cara 1

```python
# df_car[df_car['model']=='Camry']
```

```python
# buat dataframe untuk mobil Camry saja
df_camry = df_car[df_car['model']=='Camry']
df_camry
```

```python
df_camry['mileage'].mean()
```

```python
# buat kolom baru untuk menandakan apakah mobil tsb di bawah rata-rata mileage
df_camry['mileage_under'] = df_camry[['mileage']] < df_camry[['mileage']].mean()
df_camry
```

```python
df_camry[(df_camry['mileage'] < df_camry['mileage'].mean()) & (df_camry['price'] < df_camry['price'].mean())]
```

# Cara 2


```python
# cara lain
df_camry[['price']].mean()
```

```python
# buat kolom baru untuk menandakan apakah mobil tsb di bawah rata-rata harga
df_camry['price_under'] = df_camry[['price']] < df_camry[['price']].mean()
df_camry
```

```python
# Kondisi yang keduanya True
df_camry[(df_camry['mileage_under']==True) & (df_camry['price_under']==True)]
```

# **Bonus**

```python
# Tampilkan rata-rata jarak tempuh (mileage) berdasarkan tahun, 
# di mana rata-rata jarak tempuh (mileage) pada tahun tersebut di bawah median dari seluruh rata-rata jarak tempuh (mileage)
# Urutkan dari rata-rata jarak tempuh (mileage) yang paling kecil

avg_milePerYear = df_car.groupby('year').mean()[['mileage']]
avg_milePerYear[avg_milePerYear < avg_milePerYear.median()].sort_values('mileage')
```
