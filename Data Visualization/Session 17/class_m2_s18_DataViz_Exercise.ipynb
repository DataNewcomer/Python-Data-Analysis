# **No.7**

```python
# Tampilkan visualisasi yang menggambarkan perubahan jumlah rumah yang dijual dari Jan 2016 - Des 2017

# Ubah column Date dari object ke datetime
df['Date'] = pd.to_datetime(df['Date'])
df['Date']
```

```python
plt.figure(figsize=(24,8))

sns.lineplot(data=df, x='Date', y='Price', estimator=len)

plt.ylabel('Jumlah Rumah')
plt.title('Perubahan Jumlah Rumah yang Dijual Jan 2016-Des 2017')
```

```python
# ambil bulan dan tahunnya saja
df['Month'] = pd.to_datetime(df['Date']).dt.to_period('M').apply(lambda x: str(x))
df['Month']
```

```python
# Jumlah rumah yang dijual per bulan (Jan 2016 - Des 2017)
df_monthly_price = df.groupby('Month', as_index=False).count()[['Month','Price']].rename(columns={'Price':'Number of House'})
df_monthly_price
```

```python
plt.figure(figsize=(20,6))
sns.lineplot(data=df_monthly_price, x='Month', y='Number of House')
plt.xticks(rotation=90);
```

# **No.8**

```python
# Tampilkan visualisasi yang menggambarkan perbandingan dari persebaran harga rumah (Price) berdasarkan tipe rumah (Type)!
# Rumah di harga berapa untuk sebuah rumah bertipe "t" (townhouse) bisa dikatakan sebagai outlier?

sns.boxplot(x=df['Price'], y='Type', data=df)
```

```python
df_type_t = df[df['Type']=='t'][['Price']]
df_type_t
```

```python
plt.figure(figsize=(14,3))
sns.boxplot(df_type_t['Price'])
```

```python
q1 = df_type_t['Price'].quantile(0.25)
q3 = df_type_t['Price'].quantile(0.75)

IQR = q3 -q1
IQR 

upper = q3 + 1.5*IQR
lower = q1 - 1.5*IQR

print(f'Sebuah rumah bertipe "t" bisa dikatakan sebagai outlier jika harganya > {upper} atau < {lower}')
```

```python
import folium
```

```python
# Tampilkan lokasi rumah paling mahal! 
# Tandai dengan dengan icon pin warna merah, 
# Ketika cursor mouse di-hover ke lokasi akan keluar tulisan 'Rumah' dan kalau diklik akan keluar alamatnya (Address)!

df[df['Price'] == df['Price'].max()]
```

```python
# Buka peta pada lokasi dgn koordinat tertentu
rumah = folium.Map(location=[-37.93168, 145.16126], zoom_start=15)

# untuk popup
myPopup = folium.Popup('Address: 35 Bevis St', min_width=200, max_width=200)

# untuk icon
myIcon = folium.Icon(color='black', icon='info-sign')

# Kasih keterangan
folium.Marker(
    location  = [-37.93168, 145.16126], 
    popup     = myPopup,
    tooltip   = 'Rumah',
    icon      = myIcon

).add_to(rumah)

rumah
```

```python
# Andaikan kita sedang mencari apartemen (Type 'u') dengan 2 kamar tidur ('Bedroom2') yang harganya murah
# Coba tampilkan peta kota melbourne
# Berikan marker pada 5 apartemen termurah
# Beri keterangan pada popup berupa Suburb dan Alamat ('Address')
# Beri keterangan pada tooltip berupa harga ('Price')
```

```python
df_survey = df[(df['Type'] == 'u') & (df['Bedroom2'] == 2)].sort_values(by='Price').head(5).reset_index()
df_survey
```

```python
# for index, val in df_survey.iterrows():
#     print(index)
#     print(val['Lattitude'])
#     print(val['Longtitude'])
#     print(val['Suburb'])
#     print(val['Address'])
#     print(val['Price'])

#     print()
```

```python
# peta Melbourne
melb_map = folium.Map(location=[-37.8136, 144.9631], zoom_start=11)

# looping tiap baris di df_survey
for index, val in df_survey.iterrows():

    # buat marker 
    rumah_pin = folium.Marker(
        location = [val['Lattitude'], val['Longtitude']],
        popup = folium.Popup(f"Suburb: {val['Suburb']} <br> Address: {val['Address']}",
                            min_width=250,
                            max_width=250),
        tooltip = f"${val['Price']:,}",
        icon = folium.Icon(color='red', icon_color='white', icon='home')
    ) 

    # memasukkan rumah_pin pada melb_map
    rumah_pin.add_to(melb_map)

melb_map
```
