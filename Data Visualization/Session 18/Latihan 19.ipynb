# 1. Tampilkan lokasi rumah paling mahal di melbourne! 
- Tandai dengan dengan icon pin warna merah, 
- Ketika cursor mouse di-hover ke lokasi akan keluar tulisan 'Rumah' dan kalau diklik akan keluar alamatnya (Address)!

```python
df_house = pd.read_csv('melb_data.csv')
df_house.head()
```

```python
df_house = df_house.sort_values('Price', ascending=False)
df_house
```

```python
# Melbourne Map
melb_map = folium.Map(
    location=[-37.93168, 145.16126],
    zoom_start=12
)

# Marker
house_marker = folium.Marker(
    location=[-37.93168, 145.16126],
    tooltip='House',
    popup='35 Bevis St Mulgrave VIC 3170, Australia',
    icon = folium.Icon(color='red', icon='home')
)

# Memasukkan Marker kedalam Melbourne Map
melb_map.add_child(house_marker)

melb_map
    
```
