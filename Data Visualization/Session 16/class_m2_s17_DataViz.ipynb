## **Cara Menampilkan Beberapa Grafik dalam Satu Figure**

**Cara 1**

```python
# pertama: buat dulu figurenya
fig = plt.figure(figsize=(20,5))

# kedua: buat subplot (axes) yang nanti akan diisi dengan grafik
# ketiga: tulis kode untuk membuat grafik di bawah masing-masing subplot yg telah disediakan

ax1 = fig.add_subplot(131)                  # digit yg diisi dalam add_subplot adalah (baris, kolom, slot yg diisi)
sns.histplot(data=tips, x='total_bill')

ax2 = fig.add_subplot(132)
sns.boxplot(data=tips, y='total_bill')

ax3 = fig.add_subplot(133)
sns.barplot(data=tips, x='sex', y='total_bill')

plt.show()
```

**Cara 2**

```python
# Buat figure dengan axesnya. Masukkan jumlah kolom dan baris yang disediakan untuk mengisi grafik
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,5))

# Masukkan grafik pada tiap slot yang telah tersedia dengan cara mengisi parameter ax
sns.histplot(data=tips, x='total_bill', ax=axes[0])
sns.boxplot(data=tips, y='total_bill', ax=axes[1])
sns.barplot(data=tips, x='sex', y='total_bill', ax=axes[2])

plt.show()
```

```python
# unpacking axes

# Buat figure dengan axesnya. Masukkan jumlah kolom dan baris yang disediakan untuk mengisi grafik
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(20,5))

# Masukkan grafik pada tiap slot yang telah tersedia dengan cara mengisi parameter ax
sns.histplot(data=tips, x='total_bill', ax=ax1)
sns.boxplot(data=tips, y='total_bill', ax=ax2)
sns.barplot(data=tips, x='sex', y='total_bill', ax=ax3)

plt.show()
```

## Contoh Anotasi Barplot

```python
plt.figure(figsize=(8, 6))

ax = sns.barplot(x="day", y="total_bill", data=tips, ci=0)

# Iterrating over the bars one-by-one
for bar in ax.patches:

    # Using Matplotlib's annotate function and passing the coordinates where the annotation shall be done
    # x-coordinate: bar.get_x() + bar.get_width() / 2
    # y-coordinate: bar.get_height()
    # free space to be left to make graph pleasing: (0, 8)
    # ha and va stand for the horizontal and vertical alignment
	ax.annotate(format(bar.get_height(), '.2f'),
				  (bar.get_x() + bar.get_width() / 2,
					bar.get_height()), 
                    ha='center', 
                    va='center',
				    size=14, xytext=(0, 8),
				    textcoords='offset points')

plt.xlabel("Day", size=14)
plt.ylabel("Total Bill", size=14)
plt.title("Avg. Total Bill per Day")
plt.tight_layout()
plt.show()
```
